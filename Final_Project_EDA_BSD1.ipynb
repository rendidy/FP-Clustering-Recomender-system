```python
# Hitung korelasi antara kolom-kolom numerik
correlation_matrix = df3.corr()

# Plot tabel heatmap korelasi
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Tabel Heatmap Korelasi')
plt.show()
```

Di tabel Kolerasi kita mendapatkan beberapa informasi, kolom harga freight_value sangat berpengaruh oleh kolom  `price` dan `product_weight_g`.

Setelah menggabungkan data, saya ingin menentukan apakah harga mempengaruhi review score maka dari itu di tentukan lah sebuah hypothesis yaitu:

H0: Tidak adanya korelasi antara Review score dengan Harga barang/Payment Value

H1: Adanya korelasi antara Review score dengan Harga barang/Payment Value

```python
corr_r, pval_p = stats.pearsonr(df_orders_payments['review_score'], df_orders_payments['payment_value'])

print(f"r-correlation: {corr_r:.2f}, p-value: {pval_p}")
```

- Nilai r-korelasi: Koefisien korelasi Pearson (r) mengukur kekuatan dan arah hubungan linear antara dua variabel kontinu. Dalam kasus ini, nilai -0,05 menunjukkan korelasi negatif yang sangat lemah antara "review_score" dan "payment_value." Tanda negatif menunjukkan hubungan negatif, tetapi besarnya mendekati nol, yang menunjukkan bahwa hubungannya sangat lemah.

- Nilai p: Nilai p yang terkait dengan koefisien korelasi menguji hipotesis nol bahwa tidak ada korelasi antara kedua variabel tersebut. Nilai p mendekati nol (9,235464855587494e-51, yang sangat mendekati nol) menunjukkan bukti kuat menentang hipotesis nol. Dengan kata lain, ini menunjukkan bahwa ada korelasi statistik yang signifikan antara "review_score" dan "payment_value," meskipun korelasinya sangat lemah.

Secara keseluruhan, meskipun ada korelasi statistik yang signifikan antara "review_score" dan "payment_value," korelasinya sangat lemah (mendekati nol) sehingga mungkin tidak memiliki makna praktis yang signifikan. Tanda negatif menunjukkan bahwa ketika "review_score" menurun (menuju skor yang lebih rendah), "payment_value" mungkin sedikit meningkat, tetapi pengaruhnya hampir tidak terlihat maka dari itu H1 diterima.
